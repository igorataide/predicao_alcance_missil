# Modelo de predição do alcance no lançamento de um míssil.


Com base na equação do movimento oblíquo, implementaremos um modelo para fazer cálculos de alcance para os lançamentos de um míssil em diversos ângulos diferentes. Imagine o seguinte problema:

Com base na equação modelo, você deve fazer uma análise do alcance a partir dos ângulos de lançamento para as velocidades 
250 m/s, 300 m/s e 350 m/s.
Os militares podem variar a velocidade do disparo entre 250 m/s, 300 m/s e 350 m/s e o 
ângulo de 15º a 75º, variando a cada 1º. Para planejar os disparos, precisam de um dicionário 
que indique qual será o alcance com base no modelo (equação) estabelecido. Considere:

Gravidade = 9.81 m/s²

Resistência do ar = 1.589


## _LET'S DO IT._

Como se pede a distância que o míssil alcança pode-se concluir que o y da fórmula é = 0, e desta forma podemos isolar o x e encontrá- lo.

```python
# criando uma  função que para calcular a fórmula, com os parâmetros de entrada do alpha e a velocidade.
import math

def calcula_x(alfa,v0): 
    beta=1.589
    g=9.81
    x = (math.tan(math.radians(alfa))*2*(v0**2)*(math.cos(math.radians(alfa))**2))/(g*beta)
    return x
```

```python
# Para cada velocidade, e para cada ângulo dentro de cada velocidade calculando a função acima adicionando em uma lista 

velocidades =[250,300,350]
dici={}
alcances=[]

for velocidade in velocidades: #Para cada velocidade
    for angulo in range(15,75+1): # Calcular os alcances possíveis para cada ângulo.
        alcances.append({'alcance':calcula_x(angulo,velocidade),'alfa':angulo,'velocidade':velocidade})

alcances
```

O time de artilharia precisa conferir rapidamente qual zona de fogo, para isso, imprima um dicionário que contenha o alcance máximo e mínimo para cada uma das velocidades possíveis.

```python
#organizando por velocidade

from itertools import groupby

alcances = sorted(alcances, key=lambda i: i["velocidade"]) 
print(alcances)
```

```python
# Dentro da lista ordenada, pegando os máximos e mínimos de cada velocidade 
alcances_maximos=[max(alcance, key=lambda z: z["alcance"]) for k,alcance in groupby(alcances , key=lambda i: i["velocidade"])] 
alcances_minimos=[min(alcance, key=lambda z: z["alcance"]) for k,alcance in groupby(alcances , key=lambda i: i["velocidade"])] 

for dic in alcances_maximos:
    print(f"A maior distância para a velocidade {dic['velocidade']} é {dic['alcance']} metros, com um alfa de {dic['alfa']} graus.\n")

for dic in alcances_minimos:
    print(f"A menor distância para a velocidade {dic['velocidade']} é {dic['alcance']} metros, com um alfa de {dic['alfa']} graus.\n")
```

O time de artilharia realizou um teste com o disparo estabelecido a 300 m/s em uma
angulação de 30º, o alcance foi de 4500 m. Com base nesses valores, ajuste o modelo
através do parâmetro de Resistência do Ar (β).

```python
# função par o cálculo da fórmula anterior isolando beta e tendo como parâmetros a ângulo alpha, o alcance e a velocidade.

def calcula_beta(alfa,x,v0): 
    g=9.81
    beta = (math.tan(math.radians(alfa))*2*v0**2*(math.cos(math.radians(alfa))**2))/(g*x)
    return beta
```

```python
# ajustando o modelo através do parâmetro (β)

beta=calcula_beta(30,4500,300)
print(f"Com velocidade de 300 m/s, em uma angulação de 30 graus e alcance de 4500 m, a resistência do ar é {beta:.4f}.")
```

Criando um gráfico da trajetória que possui o menor alcance, dentre todas as possíveis
combinações.

```python
#Calculando valores de Y para plotar o gráfico.

from matplotlib import pyplot as plt

def calcula_y(x,alfa,v0): 
    beta=1.589
    g=9.81
    y= (math.tan(math.radians(alfa))*x)-((g*beta*x**2)/(2*v0**2*(math.cos(math.radians(alfa))**2)))
    return y
```

```python
# Pega a menor trajetória

alcance_min=min(alcances, key=lambda z: z["alcance"]) 

print(f"Dados de alcance mínimo: {alcance_min}.\n")
```

```python
# pegando o menor alcance, o alfa da menor trajetória e a velocidade da menor tragetória
legenda=[]
for dicio in alcances_minimos:
    eixox=[]
    eixoy=[]
    xf=dicio['alcance'] # pega o menor alcance
    alfa_min=dicio['alfa'] # pega o alfa da menor trajetória
    vel_min=dicio['velocidade'] #pega a velocidade da menor tragetória
    
    # Calcular eixo y para todos os x.
    
    for x in range(0,int(xf+1),1): 
        y=calcula_y(x,alfa_min,vel_min)
        eixox.append(x)
        eixoy.append(y)
        
    # plotando a curva

    plt.plot(eixox,eixoy) 
    plt.ylabel('Altura do Projétil (m).')
    plt.xlabel('Alcance do Projétil (m).')
    legenda.append(vel_min)

plt.legend(legenda,title="velocidade")

plt.show()
```
